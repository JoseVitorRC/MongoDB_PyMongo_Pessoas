```python
#Fazendo uma busca filtrada para pessoas que nasceram depois de 2005

pessoas2005 = collection.find({'Date of birth': {'$gt': '2005-01-01'}})

for pessoas in pessoas2005:
    print(pessoas)
```

```python
#Fazendo uma outra busca filtrada apenas para pessoas do sexo feminino

pessoas_feminino = collection.find({'Sex': 'Female'})

for pessoas in pessoas_feminino:
    print(pessoas)
```
